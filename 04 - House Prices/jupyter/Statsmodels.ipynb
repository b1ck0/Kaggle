```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

Reading the datasets.

```python
train_csv = pd.read_csv('../input/train.csv', index_col='Id')

train_csv.head()
```

```python
# list of columns which needs to be converted to categories
specialColumns = ['MSZoning', 'MSSubClass', 'Street', 'Alley', 'LotShape', 'LandContour', 'Utilities',
                  'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2', 'BldgType',
                  'HouseStyle', 'OverallQual', 'OverallCond', 'RoofStyle', 'RoofMatl', 'Exterior1st',
                  'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation', 'BsmtQual',
                  'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'Heating', 'HeatingQC',
                  'CentralAir', 'Electrical', 'KitchenQual', 'Functional', 'FireplaceQu', 'GarageType',
                  'GarageFinish', 'GarageQual', 'GarageCond', 'PavedDrive', 'PoolQC', 'Fence', 'MiscFeature',
                  'SaleType', 'SaleCondition'
                 ]

# Transforming the columns into a categorical type
for column in specialColumns:
    
    # handling missing data
    train_csv[column].fillna('other', inplace=True)
    
    # converting to category
    train_csv[column] = train_csv[column].astype('category')
```

```python
test_csv = pd.read_csv('../input/test.csv', index_col='Id')
test_csv.head()
```
